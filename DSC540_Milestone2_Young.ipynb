## Step 1: Removing the Duplicate "Unnamed" Column
The "Unnamed" column seems to be a **duplicate** of the index column and is unnescessary. The code below drops the column by identifying the specific column name. 

```python
#.drop() with the specified column name will remove the unwanted column
#inplace=True means that the modification is to the original DataFrame directly
Holywd_gross_df.drop(columns=['Unnamed: 0'], inplace=True)
#displaying the first 5 rows of the dataframe to verify the column was dropped
Holywd_gross_df.head()
```

## Step 2: Replace & Rearrange Headers
Some of the headers are a bit long and can be more precise. For example, several of the headers include "(in $)" for domestic sales... We can already assume that domestic sales are in U.S. dollars. For international and world wide sales, we can change it to something more professional looking such as (USD). Also, the "License" is actually displaying the movie rating, so I will change that column name to "Rating". I am also going to rearrange the headers so that similar information (such as dates) are closer together. 

```python
#replacing the header (column) names using a list
new_col_names = ['Title', 'Movie Info', 'Year', 'Distributor',
       'Budget', 'Domestic Opening', 'Domestic Sales',
       'International Sales (USD)', 'World Wide Sales (USD)', 'Release Date',
       'Genre', 'Running Time', 'Rating']
#assigning the new column names using the 'new_col_names' list
Holywd_gross_df.columns = new_col_names
```

```python
#reordering the columns to organize more relavant/similar data together
Holywd_gross_df = Holywd_gross_df[['Title', 'Year', 'Release Date', 'Genre', 'Rating',  
       'Distributor', 'Running Time', 'Movie Info', 'Budget', 'Domestic Opening', 'Domestic Sales',
       'International Sales (USD)', 'World Wide Sales (USD)']]
#displaying the first 5 rows of the dataframe to verify the column names were changed and rearranged
Holywd_gross_df.head()
```

## Step 3: Handling Outliers/Bad Data Identified in the 'Budget' Column
At index 2 the 'budget' column shows "December 14 2022 (EMEA APAC)" insead of a numerical number representing the budget in USD. After expanding the dataframe to 50 rows, it shows that there are several rows that also have similar incorrect entries for the 'budget' column. 

```python
#the pd.to_numeric function attempts to convert the values in the "Budget" column to numeric format
#the errors='coerce' parameter replaces any values that cannot be converted to numeric with NaN 
Holywd_gross_df['Budget'] = pd.to_numeric(Holywd_gross_df['Budget'], errors='coerce')
#displaying the first 20 rows of the dataframe to verify the 'Budget" column now shows NaN for non-numeric entries
Holywd_gross_df.head(20)
```

## Step 4: Finding Duplicates
Having duplicates of the same movie might make it difficult to analyze this data later on. First we will find the movies that have duplicates in the "title" column. Then we will need to investigate whether these duplicates are a duplicate entry of the data, or if there is actually more than one movie with the same name. 

```python
#creating an array of unique movie titles that are duplicated in the DataFrame
#.duplicated() method with the subset='title' parameter checks for duplicate values in the "title" column
#the keep=False parameter marks all instances of the duplicated titles as True
duplicate_titles = Holywd_gross_df[Holywd_gross_df.duplicated(subset='Title', keep=False)]['Title'].unique()
#printing the array of duplicated titles
print(duplicate_titles)
```

```python
#calculating the total number of duplicated titles using .sum()
duplicate_count = Holywd_gross_df.duplicated(subset='Title').sum()
print("The number of duplicated rows based on the 'Title' column is:", duplicate_count)
```

```python
#displaying each of the duplicated rows for further investigation...
#.isin() checks for the presence of each title in the array of duplicate_titles 
#the result is a boolean mask that selects the rows where the condition is True
duplicate_rows = Holywd_gross_df[Holywd_gross_df['Title'].isin(duplicate_titles)]
#displaying the full row for the duplicated titles
print(duplicate_rows)
```

```python
#printing a subset of the data where both the "title" and "date" columns are duplicates
duplicate_subset = Holywd_gross_df[Holywd_gross_df.duplicated(subset=['Title', 'Year'], keep=False)]
print(duplicate_subset)
#the results are an empty dataframe, which means that NONE of the rows are a duplicate entry of the data
```

## Step 5: Editing Duplicate Movie Titles
Based on the results above, none of the rows are exact duplicate entries of the data. If there were duplicate entries of the data, we could remove those rows using the .drop() method. However, it seems that our duplicate movie titles are movies that have been remade years later. In this instance one option is to identify the remake movies with "(Remake)" added to the end of the title. 

```python
#creating a Series 'duplicate_titles' containing the titles that are duplicated in the 'Title' column
duplicate_titles = Holywd_gross_df[Holywd_gross_df.duplicated(subset='Title', keep=False)]['Title']
#filters so that the series only includes Series the titles that are duplicated and ignores the first occurance of the title
titles_to_rename = duplicate_titles[duplicate_titles.duplicated(keep='first')]

#a loop that iterates through each title in the titles_to_rename
for title in titles_to_rename:
    #the code retrieves the indices of the rows where the 'Title' column matches the current title
    indices = Holywd_gross_df[Holywd_gross_df['Title'] == title].index
    #a loop that for each index in the indices list, starting from the second index (index 1) 
    for i in range(1, len(indices)):
        #the code updates the 'Title' column of the DataFrame at the specified index by appending "(Remake)" to the current title
        Holywd_gross_df.at[indices[i], 'Title'] = f"{title} (Remake)"
```

## Step 6: Editing Monetary Values into a More Readable Format
Currently the monetary values for the 'Budget', 'Domestic Opening', 'Domestic Sales', 'International Sales (USD)', and 'World Wide Sales (USD)' columns are displayed as a long list of numbers such as: 1453683476. In order to change the monetary values to a more readable format I will need to first convert each of the columns to a numeric format, then format the numeric values to US currency. I will need to be conscious that the "Budget" column contains NaN values due to removing bad data earlier, so I will have to tell the code to "ignore" those NaN values. 

```python
#importing the locale module which can be used to format numbers and currencies
import locale

#setting the locale to the United States (since our monetary values are in USD)
#the LC_ALL constant represents all categories of the locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

#create a list of columns to format
columns_to_format = ['Budget', 'Domestic Opening', 'Domestic Sales', 'International Sales (USD)', 'World Wide Sales (USD)']

#converting the columns to numeric format and formatting the values
#a for loop that iterates through each column in the columns_to_format list
for col in columns_to_format:
    #converts the column values to numeric using pd.to_numeric(), with errors coerced to NaN
    Holywd_gross_df[col] = pd.to_numeric(Holywd_gross_df[col], errors='coerce')
    #applying a lambda function to format the numeric values as currency using the locale.currency
    #the grouping parameter set to True to include the grouping separator
    #if a NaN value is encountered, it is ignored
    Holywd_gross_df[col] = Holywd_gross_df[col].apply(lambda x: locale.currency(x, grouping=True) if not np.isnan(x) else np.nan)
```
